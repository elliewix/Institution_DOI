DataCite Data File 

Using DataCite's API, I collected metadata from institutions that had contributed datasets to DataCite with a publication year of 2017. This data was collected with my DataCite_API and DataCite_API_pages(41-100).py scripts. After I collected the data, I ran it through DataCite_csvconvert.py script to extract what I wanted and convert it to a csv file. I then used my DataCite_agg.py to group by all of the rows in DataCite_doi.csv by institution name. In my DataCite_agg.py script, I removed a lot of odd data and listed it as Missing URL, but the count of missing url's did not transfer to my new dataset, so I handcoded the number, 274.

```python
DataCitedata = pd.read_csv("curated_DataCite_doi_2.0.csv") # read in the DataCite csv file
DataCitedata.index = DataCitedata.index + 1  # start index at 1
DataCitedata = DataCitedata.drop(columns=['Unnamed: 0']) # clean messy data
DataCitedata = DataCitedata.drop(columns=['institution code']) # replaced by hand coding instituion names
```

DataVerse Data File

I created this csv file from DataVerse publication metadata on their browsing digital library feature from institutions that contributed the most datasets to Dataverse in the 2017 publication year. I selected only datasets for the 2017 publication, and then copy pasted those results into a DataVerse_doi.csv. I then created a new a file called curated_DataVerse_doi.csv where I did some cleaning and curation. I edited King's College London from a zero value to ten for two reasons. First, the institution appeared in order by decreasing frequency with the other ten counts. Secondly, King's College London had ten DOI's for 2017. By hand I merged MIT and Massachusetts Institute of Technology institutions into just MIT. There were also two Harvard T.H. Chan School of Public Health rows; I combined these entities  because I looked at DataVerse recently and there is now only one occurence of Harvard T.H. Chan School of Public Health with a similar DOI count. I also merged ICRAF, World Agroforestry Center, and World Agroforestry (ICRAF) because these are all the same entity. There were also two Stanford entities, so I merged them as well. Additionally, I deleted the Northeastern University / Harvard University row because I could not ascertain what institution it belonged to. Lastly, I added a new column Data Source and put Harvard DataVerse for each row. 

```python
DataVersedata = pd.read_csv("curated_DataVerse_doi.csv") # read in DataVerse data file
DataVersedata.index = DataVersedata.index + 1  # start index at 1
```

OpenAire Data File Similar to DataVerse, I copy pasted publication metadata from institutions that have contributed the most datasets to OpenAire for the 2017 publication year. After which I hand coded OpenAire_doi.csv. I then made a new file called curated_OpenAire_doi.csv where I added the Data Source Column and gave each row an OpenAire value under this column. I also deleted the Unknown Repository entity. Since CLARIN.SI repository and LINDAT/CLARIN repository are the same entity, I merged them.

```python
OpenAiredata = pd.read_csv("curated_OpenAire_doi.csv") # read in OpenAire data file
OpenAiredata.index = OpenAiredata.index + 1 # reset index
```

GRID Data File

The data in this file comes from two datasets in the GRID database, grid.csv and types.csv. Using pandas, I performed a left merge on the two datasets on grid id and id because I wanted to add institution type to grid.csv file. Having merged the two files together, I am hand coding a new csv file, grid_curated.csv, which will only include institutions from my data. I am doing this by using ctrl f to find said institution and copy paste it into grid_curated.csv. 

This file does not often differentiate  between a universitiy's library and the institution itself, so I would use the institution information for a library from my data, but I replaced the type value of Education with Archive because that is the GRID value assigned to libraries. 

I also put filled other for institution type for some of the very large DOI producing data repositories that were not represented in GRID's Database (Harvard IQSS Dataverse, ICPSR, Figshare, GTEx, Pangaea, Research Gate, TOPMed, and Zenodo). I chose other because this is how GRID usually classified a data repository
